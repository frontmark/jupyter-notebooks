# Interest Rate Converter

This tool converts continuously compounded interest rates in interest rates with arbitrary compounding and vice versa.

```python
import math
```

```python
def CompConverter(_type, m, r):
    if _type == 'c': #continuously required
        r = m*math.log(1+r/m)
    if _type == 'nc':# non continuously required
        r = m*(math.exp(r/m)-1)
    return print(r)
```

In order to convert interest rates, you need to provide the formula the following arguments:

- desired type of compounding (continuously ('c') or non-continuously ('nc')
- compounding frequency $m$
- interest rate $r$

```python
rate = CompConverter('c',2,0.2)
```
